```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
sns.set()
%matplotlib inline
```

```python
#distance of vectors from origin.
dims = [1, 1000]

def norms(x):
    return np.sqrt(np.sum(np.square(x), axis=1))

# Distribution of ||x||
plt.figure()
for d in dims:
    x = np.random.randn(1000, d)
    sns.distplot(norms(x))#, kde_kws={'label' : 'dim={0:d}'.format(d)})
    plt.xlabel('Norm')
plt.show()
```

```python
s = np.random.uniform(1,-1,(401,1000))
print(s.shape)
print(s[0].shape)
```

```python
#distances between vectors
save=s[400]
diff=[]
for i in range(400):
    diff.append(s[i]-save)
diff=np.asarray(diff)
print(diff.shape)
lengths=norms(diff)
sns.distplot(lengths)
#print(lengths[200])
#count, bins, ignored = plt.hist(diff, 15, density=True)
#plt.plot(bins, np.ones_like(bins), linewidth=2, color='r')
#plt.show()
```

```python
#cosine of the angles between vectors
angles=[]
normsave=np.linalg.norm(save)
for i in range(400):
    angles.append(np.dot(s[i],save)/normsave/np.linalg.norm(s[i]))
angles=np.asarray(angles)
sns.distplot(angles)
```
